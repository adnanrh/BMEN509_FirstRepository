```python
# Library Imports
import os
import math
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from scipy import signal, ndimage

# Setup Jupyter notebook specifics
%matplotlib inline

# Numpy printing help
np.set_printoptions(precision=2)
```

```python
Im_size = (100, 100)

Im_a = np.zeros(shape=Im_size)
for i in range(Im_size[0]):
    for j in range(Im_size[1]):
        Im_a[i, j] = 20*np.sin(2*np.pi*i) + 20*np.sin(2*np.pi*j)

plt.subplots(1, 2, figsize=(10,5))
plt.subplot(1, 2, 1); plt.imshow(Im_a, cmap='gray'); 
plt.subplot(1, 2, 2); plt.hist(Im_a.ravel(), bins=256); plt.xlabel('Intensity'); plt.ylabel('Count')
plt.show()

Im_b = np.zeros(shape=Im_size)
for i in range(Im_size[0]):
    for j in range(Im_size[1]):
        Im_b[i, j] = 2*np.exp(-((((i-5)**2)/2) + ((j-5)**2)/2))
        
plt.subplots(1, 2, figsize=(10,5))
plt.subplot(1, 2, 1); plt.imshow(Im_b, cmap='gray'); 
plt.subplot(1, 2, 2); plt.hist(Im_b.ravel(), bins=256); plt.xlabel('Intensity'); plt.ylabel('Count')
plt.show()

Im_c = np.zeros(shape=Im_size)
for i in range(Im_size[0]):
    for j in range(Im_size[1]):
        Im_c[i, j] = 20*np.sin(2*np.pi*i)
        
plt.subplots(1, 2, figsize=(10,5))
plt.subplot(1, 2, 1); plt.imshow(Im_c, cmap='gray'); 
plt.subplot(1, 2, 2); plt.hist(Im_c.ravel(), bins=256); plt.xlabel('Intensity'); plt.ylabel('Count')
plt.show()

Im_d = np.zeros(shape=Im_size)
for i in range(Im_size[0]):
    for j in range(Im_size[1]):
        Im_d[i, j] = 1 if (i > -10 and i < 10 and j > -10 and j < 10) else 0
        
plt.subplots(1, 2, figsize=(10,5))
plt.subplot(1, 2, 1); plt.imshow(Im_d, cmap='gray'); 
plt.subplot(1, 2, 2); plt.hist(Im_d.ravel(), bins=256); plt.xlabel('Intensity'); plt.ylabel('Count')
plt.show()
```
